## Peak finding

Write a function, `find_peaks`, that finds and returns the indices of the *local maxima* in a sequence of values in an array. A *local maxima* is defined as a value whose neighboring values are all smaller. Your function should:

* Properly handle local maxima at the endpoints of the input array.
* Return a Numpy array of integer indices.
* Handle any Python iterable as input.

def find_indicies(a):
    for i in range(0, len(a)):
        if a[i] > a[i+1] and a[i] > a[i-1]:
            yield i

```python
def rollRight(a):
    aRight = [0]
    for i in a[0:-1]:
        aRight.append(i)
    return aRight
```

```python
def rollLeft(a):
    aLeft = []
    for i in a[1:]:
        aLeft.append(i)
    aLeft.append(0)
    return aLeft
```

```python
def find_peaks(a):
    """Find the indices of the local maxima in a sequence."""
    if type(a) is not list:
        a = list(a)
    L = [a]
    L.append(rollRight(a))
    L.append(rollLeft(a))
    arrL = [np.array(i) for i in L]
    arrL.append(arrL[0] - arrL[1])
    arrL.append(arrL[0] - arrL[2])
    indexL = [list(np.where(arrL[3] > 0))] 
    indexL.append(list(np.where(arrL[4] > 0)))   
    l0 = len(indexL[0][0])
    l1 = len(indexL[1][0])
    indexArrL = [np.array(i) for i in indexL]
    if l0 > l1:
        indexL[1][0] = np.pad(indexL[1][0], (l0 - l1)//2, 'constant', constant_values=-1)
    elif l1 > l0:
        indexL[0][0] = np.pad(indexL[0][0], (l1 - l0)//2, 'constant', constant_values=-1)
    indicies = np.intersect1d(indexArrL[0][0], indexArrL[1][0])
    return indicies
```

```python
p1 = find_peaks([2, 0, 1, 0, 2, 0, 1])
p1
```

```python
p2 = find_peaks([0,1,2,3])
p2
```

```python
p1 = find_peaks([2,0,1,0,2,0,1])
assert np.allclose(p1, np.array([0,2,4,6]))
p2 = find_peaks(np.array([0,1,2,3]))
assert np.allclose(p2, np.array([3]))
p3 = find_peaks([3,2,1,0])
assert np.allclose(p3, np.array([0]))
```

Here is a string with the first 10000 digits of $\pi$ (after the decimal):

```python
from sympy import pi, N
pi_digits_str = str(N(pi, 10001))[2:]
print(pi_digits_str[0:100]+'...')
```

Write code to perform the following:

* Convert that string to a Numpy array of integers.
* Find the indices of the local maxima in the digits of $\pi$.
* Use `np.diff` to find the distances between consequtive local maxima.
* Save the result in a Pandas `DataFrame` with a column name of `Distance`.

```python
piArr = np.fromiter(pi_digits_str, dtype=np.int)
piPeaks = find_peaks(piArr)
piDict = {'Distance':np.diff(piPeaks)}
df = pd.DataFrame(piDict)
```

```python
assert type(df)==pd.DataFrame
assert df.columns[0]=='Distance'
assert len(df)==2898
assert df['Distance'].dtype==int
```
