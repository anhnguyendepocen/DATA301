## Question 1

Create as many cells as you need below this one to explore propose, explore and answer your first question.

Question 1: Are the primary types of pokemon balanced? That is, does the Total of a certain type of pokemon far greater than others?

Let's start by figuring out how many pokemon of each type there are.

```python
Chart(data).mark_bar().encode(
    X('Type 1:O'),
    Y('count(*):Q', title='Number of Pokemon')
)
```

Now, let's see if the bar graphs of the summed Total match the general shape of the number of pokemon.

```python
Chart(data).mark_bar().encode(
    X('Type 1:O'),
    Y('sum(Total):Q')
)
```

Now, we'll average these totals in order to answer the question.

```python
Chart(data).mark_bar().encode(
    X('Type 1:O'),
    Y('average(Total):Q', title='Averaged Total')
)
```

Answer: It would appear that the game designers kept balance in mind by keeping a type's average total stats around 450. However, based on the graph above Dragon type pokemon tend to be more powerful overall than any other type. Note that this exploration does not account for the secondary type of a pokemon.

Question: How many of each primary type is paired with a secondary type? Which primary type has the most pairs with a secondary type?

```python
data1 = data.dropna()
```

```python
Chart(data1).mark_bar(opacity=0.5).encode(
    X('Type 1', sort=SortField(field='count(*)', order='ascending')),
    Y('Type 2'),
    Color('count(*)', scale=Scale(range=['yellow', 'purple']))
)
```

Answer: Grass, Fire, and Bug primary types are most paired with a secondary type. Flying, Fighting, and Steel types are least paired with any secondary type. Normal/Flying pokemon are the most commonly paired types in the game.
