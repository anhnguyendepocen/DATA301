The only module used for visualization was *Altair*.

```python
from altair import *
import pandas as pd
```

**Transformed Files**

```python
savedFileDir = '/data/chazalex/transforms/' #file directory from which to load transformed files
savedFileNames = ['spPerceptions.csv', 'faSpPerceptions.csv', 'battery.csv', 'batteryPartIDAvg.csv', 'batteryDateAvg.csv', 
                  'batMergeSpDate.csv']
dfList = [pd.read_csv('/data/chazalex/transforms/'+fileName) for fileName in savedFileNames]
```
